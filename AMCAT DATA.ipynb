### Step 3: Univariate Analysis
Histograms, Boxplots, Countplots, and Outliers


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Univariate Analysis - Salary Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Salary'].dropna(), kde=True, bins=30, color='blue')
plt.title('Salary Distribution with KDE')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Boxplot for Salary to detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Salary'].dropna(), color='orange')
plt.title('Boxplot of Salary')
plt.show()

# Countplot for Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data, palette='Set2')
plt.title('Count of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Histogram for 10th Percentage
plt.figure(figsize=(10, 6))
sns.histplot(data['10percentage'].dropna(), kde=True, bins=30, color='green')
plt.title('10th Grade Percentage Distribution')
plt.xlabel('10th Percentage')
plt.ylabel('Frequency')
plt.show()

# Boxplot for 10th Percentage
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['10percentage'].dropna(), color='purple')
plt.title('Boxplot of 10th Percentage')
plt.show()
```

### Step 4: Bivariate Analysis
Scatter Plots, Hexbin Plots, Pair Plots, etc.


```python
# Scatter plot to discover relationships between numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='10percentage', y='Salary', data=data)
plt.title('Relationship between 10th Percentage and Salary')
plt.xlabel('10th Percentage')
plt.ylabel('Salary')
plt.show()

# Hexbin plot between 12th Percentage and Salary
plt.figure(figsize=(10, 6))
plt.hexbin(x=data['12percentage'], y=data['Salary'], gridsize=30, cmap='Blues')
plt.colorbar()
plt.title('Hexbin plot of 12th Percentage vs Salary')
plt.xlabel('12th Percentage')
plt.ylabel('Salary')
plt.show()

# Pair plot for visualizing relationships between all numerical columns
sns.pairplot(data[['Salary', '10percentage', '12percentage', 'conscientiousness', 'agreeableness']])
plt.show()

# Categorical vs Numerical columns: Swarmplot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Gender', y='Salary', data=data, palette='Set2')
plt.title('Swarmplot of Gender vs Salary')
plt.show()

# Boxplot for Specialization vs Salary
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='10percentage', data=data, palette='coolwarm')
plt.title('Boxplot of Gender vs 10th Percentage')
plt.show()
```

### Step 5: Research Questions
1. Testing the claim from the Times of India article


```python
# Check salary data for the mentioned job roles
roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
filtered_data = data[data['Designation'].isin(roles)]

# Calculate mean salary for fresh graduates
average_salary = filtered_data['Salary'].mean()

print(f"Average salary for fresh graduates in these roles: {average_salary:.2f} Lakhs")
```

### 2. Relationship between Gender and Specialization


```python
# Cross-tabulation of Gender and Specialization
gender_specialization = pd.crosstab(data['Gender'], data['JobCity'])

# Stacked bar plot
gender_specialization.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender vs Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Job Role Preferences by City:

You could analyze which cities employ the highest number of people for specific job roles. For example, which cities have more software engineers, and is there a geographical preference for certain roles?




```python
# Group data by JobCity and Designation
city_role_data = data.groupby(['JobCity', 'Designation']).size().unstack()

# Stacked bar plot for JobCity vs Job Roles
city_role_data.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='coolwarm')
plt.title('Job Roles Distribution Across Cities')
plt.xlabel('Job City')
plt.ylabel('Count')
plt.show()
```

### Correlation Between Personality Traits and Salary:

Investigate if certain personality traits (like conscientiousness or agreeableness) have any impact on salary or job role. A positive correlation may indicate that certain traits are valued in high-paying jobs.


```python
# Correlation matrix for personality traits and salary
personality_cols = ['conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']
correlation_data = data[personality_cols + ['Salary']].corr()

# Heatmap of correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Personality Traits and Salary')
plt.show()
```

### Career Growth Analysis:

Analyze whether individuals who have stayed longer with a company (based on the difference between DOJ and DOL) have a higher salary than those with shorter tenures.

```python
# Convert DOJ and DOL to datetime format
data['DOJ'] = pd.to_datetime(data['DOJ'], errors='coerce')
data['DOL'] = pd.to_datetime(data['DOL'], errors='coerce')

# Calculate tenure (in days) and add to the dataset
data['Tenure'] = (data['DOL'] - data['DOJ']).dt.days

# Scatter plot of tenure vs salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Tenure', y='Salary', data=data)
plt.title('Tenure vs Salary')
plt.xlabel('Tenure (Days)')
plt.ylabel('Salary')
plt.show()
```
